```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the 'Cars' dataset
data = pd.read_csv("Cars.csv")
```

```python
def desc_stats(data):  
    # Convert the dataset to a Pandas DataFrame
    Cars = pd.DataFrame(data)

    # Display a brief summary of the dataset
    print(Cars.info())

    # Glimpse the first few rows of the dataset
    print(Cars.head())

    # Summary of the numeric variables
    print(Cars.describe())
    
    return None
```

```python
def scatter_mpg(data):
    # Create the scatter plot
    Cars=data
    
    plt.figure(figsize=(5, 5))
    sns.scatterplot(data = Cars, x = 'weight', y = 'mpg', palette='gray')

    # Set labels for the axes
    plt.xlabel('Car weight (lbs)')
    plt.ylabel('Miles per gallon (MPG)')
    plt.title('Relation between Car Weight and MPG')

    # Show the plot
    plt.show()
    return None
```

```python
def scatter_acc(data):
    Cars=data
    
    # Create the scatter plot
    plt.figure(figsize=(5, 5))
    sns.scatterplot(data = Cars, x = 'weight', y = 'acceleration', palette='gray')

    # Set labels for the axes
    plt.xlabel('Car weight (lbs)')
    plt.ylabel('Acceleration')
    plt.title('Relation between Car Weight and Acceleration')

    # Show the plot
    plt.show()
    return None
```

```python
def fitted_acc(data):
    Cars=data
    
    # Create the fitted plot
    plt.figure(figsize=(5, 5))
    sns.lmplot(data=Cars, x='weight', y='acceleration', height=4, aspect=1, palette='gray')

    # Set labels for the axes
    plt.xlabel('Car weight (lbs)')
    plt.ylabel('Acceleration')
    plt.title('Relation between Weight and Acceleration')

    # Show the plot
    plt.show()
    
    return None
```

```python
#NBVAL_CHECK_OUTPUT
desc_stats(data)
```

```python
# NBVAL_IGNORE_OUTPUT
scatter_mpg(data)
```

```python
# NBVAL_IGNORE_OUTPUT
scatter_acc(data)
```

```python
# NBVAL_IGNORE_OUTPUT
fitted_acc(data)
```
